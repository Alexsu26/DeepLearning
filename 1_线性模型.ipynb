```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4]
y = [2, 4, 6, 8]
```

```python
def forward(w, x) :
    return w * x

def mse(w, x, y):
    return (y - forward(w, x)) * (y - forward(w, x))
```

```python
w_list = []
mse_list = []
for w in np.arange(0, 4.1, 0.1) :
    mse_sum = 0
    for x_i, y_i in zip(x, y):
        mse_sum += mse(w, x_i, y_i)
    mse_list.append(mse_sum / len(x))
    w_list.append(w)
```

```python
plt.plot(w_list, mse_list)
plt.xlabel("w")
plt.ylabel("loss")
plt.show()
```
